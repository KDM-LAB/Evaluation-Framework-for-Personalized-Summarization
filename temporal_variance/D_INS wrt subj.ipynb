# JSD

```python
import numpy as np
import scipy as sp

def jsd(p, q, base=np.e):
    '''
        Implementation of pairwise `jsd` based on  
        https://en.wikipedia.org/wiki/Jensen%E2%80%93Shannon_divergence
    '''
    
    ## convert to np.array
    p_values, q_values = np.array(list(p.values())), np.array(list(q.values()))
    
    ## normalize p, q to probabilities
    p_arr, q_arr = p_values/p_values.sum(), q_values/q_values.sum()
    m = 1./2*(p_arr + q_arr)
    return sp.stats.entropy(p_arr,m, base=base)/2. +  sp.stats.entropy(q_arr, m, base=base)/2.
```

# Deviation

```python
def dev(uid, model_summ_dict, base_summ_dist):
    dev_sum = 0
    for user in model_summ_dict:
        if user != uid:
            dev_sum += ((jsd(model_summ_dict[uid], model_summ_dict[user]) * jsd(model_summ_dict[uid], model_summ_dict[user]))/jsd(model_summ_dict[uid], base_summ_dist))
    return dev_sum
```

# Vanilla D-INS

```python
dataset = data

vanilla_sum = 0
for key in dataset:
    dev_sum = 0
    model_summ_dict = dataset[key]['m_dict']['1'] #dictionary of { user_id: model_summary_distribution}
    print('-----------'+key+'-------------')
    for uid in model_summ_dict:
        x = dev(uid, model_summ_dict, dataset[key]['doc_text'])
        dev_sum += x
        print(x)
    dev_sum /= len(model_summ_dict)
    
    vanilla_sum += dev_sum
    
    print('deviation: ',dev_sum)
    print('vanilla_sum: ', vanilla_sum)

vanilla_sum /= len(dataset)
vanilla_DINS = 1 - vanilla_sum
```

```python
print(vanilla_sum)
print(vanilla_DINS)
```
